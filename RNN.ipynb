```python
# function to convert the words of the unzipped files into dictionary
# file will have a words with n (n can be 50 or 100 or 200 or 300) numbers

import numpy as np
def add_to_dict(filename,d): # d in the dictionary
    with open(filename,'r') as f:
        for line in f.readlines():
            l=line.split()
            d[l[0]]=np.array(l[1:],dtype='float')
```

```python
# STEPS OF NLP 
# 1) paragrap to sentence
# 2) sentence to words
# 3) convert words into base froms
# 4) reomve stop words 
# 5)
```

```python
from nltk.stem import WordNetLemmatizer # to converts verbs into their base forms --> done by lemmatizer
Lem=WordNetLemmatizer()
```

```python
def msg_to_words(s,words): # take a sentence as input and divide into tokens and returns a list of tokens(words)
    tokens=nltk.word_tokenize(s) # tokens will have list of words 
    lower_case=[t.lower() for t in tokens]

    # apply lemmatizer to convert words into base to stem from
    lem_words=[Lem.lemmatize(t) for t in lower_case]

    # remove the stopping words
    # check if a lemmatize word is not in dictinoary, then remove it otherwise return
    useful_tokens=[t for t in lem_words if t in words]
    return useful_tokens
```

```python
msg_to_words('i am so Happy at Sistech',words)
```

```python
#  to convert the list of words into feature vector using word embedding in the dictinary words
def msg_to_vector(s,words):
    # call msg_to_words method for the stemming and tokanization
    useful_tokens=msg_to_words(s,words)
    vector=[]
    for t in useful_tokens:
        if t not in words:
            continue
        vector.append(words[t])
    return np.array(vector,dtype='float')
```

```python
# to add padding to the sentences
def pad(x,length): # length is the length of the longest sentence
    x_copy=[]
    for i in x:
        diff=length-i.shape[0]
        pad=np.zeros(shape=(diff,50))
        b=np.concatenate([i,pad])
        x_copy.append(b)
    return np.array(x_copy)
```

```python
def df_to_x_y(df,words):
    x=[]
    y=df['sentiment'].to_numpy().astype('int')
    for msg in df['reivew']:
        msg_in_matrix=msg_to_vector(msg,words)
        x.append(msg_in_matrix)
    return x,y
```

```python
x,y=df_to_x_y(df,words)
```

```python
xarr=pad(x,7)
```

```python
xarr.shape
```

```python
# now make the model
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout,Input
model=Sequential()
model.add(Input(shape=(7,50))) # for one sentence
model.add(LSTM(64,activation='relu',return_sequences=True,input_shape=(3,1))) # specify input_shape as 2d shape 
model.add(Dropout(0.2))
model.add(LSTM(64,activation='relu',return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) # output is not in onehot encoding, so use sparse_catergorical_crossentropy
model.fit(xarr,y,epochs=10)
```
